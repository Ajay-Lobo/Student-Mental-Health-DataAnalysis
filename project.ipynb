```python
import csv
import pandas as pd

# Read the existing CSV file into a DataFrame
df = pd.read_csv("Student_Mental_health.csv")

# Data to append
data_to_add = ["15/10/2023 09:00", "Female", 24, "Electrical Engineering", "2nd Year", 3.6, "Single", "No", "Yes", "Yes", "Yes"]

# Create a new DataFrame from the data to add
new_row = pd.DataFrame([data_to_add], columns=df.columns)

# Concatenate the original DataFrame and the new row
df = pd.concat([df, new_row], ignore_index=True)

# Write the updated DataFrame back to the CSV file
df.to_csv("Student_Mental_health.csv", index=False)
```
